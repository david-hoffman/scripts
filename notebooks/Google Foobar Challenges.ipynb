````python
def answer(n):
    '''
    returns the smallest positive integer base b, at least 2, in which the integer n
    is a palindrome. The input n will satisfy "0 <= n <= 1000."

    Parameters
    ----------
    n : int
        The integer to test

    Returns
    -------
    b : int
        the base in which n is a palindrome

    Examples
    --------
    >>> answer(0)
    2
    >>> answer(42)
    4
    '''

    palindrome = False

    b = 1

    while not palindrome:
        b += 1
        m = newbase(n, b)
        palindrome = (m == m[::-1])

    return b

def newbase(n, b):
    '''
    Transforms a n from decimal to base b

    Parameters
    ----------
    n : int
        number to transform
    b : int
        new base

    Returns
    -------
    m : list
        list of coefficients in new base

    Examples
    --------
    >>> newbase(15, 3)
    [1, 2, 0]
    '''

    toreturn = []
    r = n
    while r > 0:
        toreturn.append(r % b)    
        r = r // b

    return toreturn[::-1]
````

Peculiar balance
================

Can we save them? Beta Rabbit is trying to break into a lab that contains the only known zombie cure - but there's an obstacle. The door will only open if a challenge is solved correctly. To help Beta get into the room, write a method called answer(x), which outputs a list of strings representing where the weights should be placed, in order for the two sides to be balanced, assuming that weight on the left has mass x units.

The first element of the output list should correspond to the 1-unit weight, the second element to the 3-unit weight, and so on. Each string is one of: 

"L" : put weight on left-hand side 
"R" : put weight on right-hand side 
"-" : do not use weight 

To ensure that the output is the smallest possible, the last element of the list must not be "-".

x will always be a positive integer, no larger than 1000000000.

````python
def newbase(n, b):
    '''
    Transforms a n from decimal to base b

    Parameters
    ----------
    n : int
        number to transform
    b : int
        new base

    Returns
    -------
    m : list
        list of coefficients in new base

    Examples
    --------
    >>> newbase(15, 3)
    [1, 2, 0]
    '''

    toreturn = []
    r = n
    while r > 0:
        toreturn.append(r % b)    
        r = r // b
    if toreturn == []:
        toreturn = [0]

    return toreturn[::-1]

def to_decimal(n, b):
    '''
    Converts a number from base b to decimal
    
    Parameters
    ----------
    n : list
        list of coefficients in base b
    b : int
        the base system in use
        
    Returns
    -------
    d : int
        the integer as a decimal number
        
    Examples
    --------
    >>> to_decimal([1, 0, 1], 3)
    10
    '''
    toreturn = 0
    for i, x in enumerate(n[::-1]):
        toreturn += x*b**i
    
    return toreturn

def one_or_zero(l):
    '''
    A utility function to check if all coefficients are 1 or zero
    '''
    for i in l:
        if i != 0 and i != 1:
            return False
        
    return True

def num_to_add(num, b):
    '''
    A utility function to determine the number, x, which when added to num
    will increment the first nonzero or unity bit
    '''
    toreturn = []
    for i in newbase(num, b)[::-1]:
        if i == 1 or i == 0:
            toreturn.append(0)
        else:
            toreturn.append(b - i)
            break
    return to_decimal(toreturn[::-1], b)

def incrementor(n, b):
    '''
    Increments n to the next number that has only 1's or 0's as each digit
    '''
    i = 1
    while i:
        i = num_to_add(n, b)
        n += i
    
    return n

def answer(x):
    if one_or_zero(newbase(x,3)):
        right = x
        left = 0
    else:
        right = incrementor(x, 3) # to add to right side
        left = right - x # to add to left side
        
    right_3 = (newbase(right, 3))[::-1]
    left_3 = (newbase(left, 3))[::-1]
    
    lenl = len(left_3)
    lenr = len(right_3)
    
    assert lenl < lenr
    
    # pad
    left_3 += [0]*(lenr - lenl)
    
    toreturn = []
    for l, r in zip(left_3, right_3):
        assert l + r < 2
        if l:
            toreturn.append("L")
        elif r:
            toreturn.append("R")
        else:
            toreturn.append("-")
            
    return toreturn
````

Zombit infection
================

Dr. Boolean continues to perform diabolical studies on your fellow rabbit kin, and not all of it is taking place in the lab. Reports say the mad doctor has his eye on infecting a rabbit in a local village with a virus that transforms rabbits into zombits (zombie-rabbits)!

Professor Boolean is confident in the virus's ability to spread, and he will only infect a single rabbit. Unfortunately, you and your fellow resistance agents have no idea which rabbit will be targeted. You've been asked to predict how the infection would spread if uncontained, so you decide to create a simulation experiment. In this simulation, the rabbit that Dr. Boolean will initially infect will be called "Patient Z".

So far, the lab experts have discovered that all rabbits contain a property they call "Resistance", which is capable of fighting against the infection. The virus has a particular "Strength" which Dr. Boolean needs to make at least as large as the rabbits' Resistance for it to infect them. 

You will be provided with the following information:
population = A 2D non-empty array of positive integers. (The dimensions of the array are not necessarily equal.) Each cell represents one rabbit, and the value of the cell represents that rabbit's Resistance. All cells contain a rabbit.
x = The X-Coordinate (column) of "Patient Z" in the population array.
y = The Y-Coordinate (row) of "Patient Z" in the population array.
strength = A constant integer value representing the Strength of the virus.

Here are the rules of the simulation: First, the virus will attempt to infect Patient Z. Patient Z will only be infected if the infection's Strength equals or exceeds Patient Z's Resistance. From then on, any infected rabbits will attempt to infect any uninfected neighbors (cells that are directly - not diagonally - adjacent in the array). They will succeed in infecting any neighbors with a Resistance lower than or equal to the infection's Strength. This will continue until no further infections are possible (i.e., every uninfected rabbit adjacent to an infected rabbit has a Resistance greater than the infection's Strength.)

You will write a function answer(population, x, y, strength), which outputs a copy of the input array representing the state of the population at the end of the simulation, in which any infected cells value has been replaced with -1.
The Strength and Resistance values will be between 0 and 10000. The population grid will be at least 2x2 and no larger than 50x50. The x and y values will be valid indices in the population arrays, with numbering beginning from 0.


````python
class simulation(object):
    '''
    A class to run the simulation
    '''
    
    def __init__(self, population, x, y, strength):
        '''
        Simulation object
        
        Parameters
        ----------
        population : two dimensional array
            the initial population of rabbits
        x : int
            x location of patient zero
        y : int
            y location of patient zero
        strength : int
            strength of the virus
        '''
        self.initial = population
        # make a copy of the initial for the current
        self.current = population[:]
        self.sizex = len(population[0])
        self.sizey = len(population)
        self.strength = strength
        # Check if the initial rabbit is susceptible
        if self.initial[y][x] <= self.strength:
            # if so, infect
            self.current[y][x] = -1
            # run simulation
            self.run()
        # if not, then the simulation is done!
    
    @staticmethod
    def print_array(a):
        '''
        Utility function to print arrays
        '''
        for b in a:
            for c in b:
                print(str(c).center(6), end='\t')
            print('')
        
    def print_current(self):
        '''
        Print current array
        '''
        print_array(self.current)

    def print_initial(self):
        '''
        Print initial array
        '''
        print_array(self.initial)
        
    def update_cell(self, x, y):
        '''
        Update a single cell located at y, x
        '''
        
        # generate the cells to check in both directions
        y_cells_to_check = [[y, i] for i in range(x-1, x+2)]
        x_cells_to_check = [[j, x] for j in range(y-1, y+2)]
        # loop through coordinates checking if the new coordinate needs to be infected
        for yy, xx in (y_cells_to_check + x_cells_to_check):
            # check validity of indices
            if xx > -1 and yy > -1 and xx < self.sizex and yy < self.sizey:
                # if cell at yy, xx is susceptible, infect
                if self.current[yy][xx] <= self.strength:
                    self.current[yy][xx] = -1
    
    def update(self):
        '''
        update the whole current array
        '''
        # loop through array
        for y, sub_array in enumerate(self.current):
            for x, resistance in enumerate(sub_array):
                # if current location is infected, update surrounding cells
                if resistance < 0:
                    self.update_cell(x, y)
                    
    def run(self):
        '''
        Run the simulation
        '''
        
        done = False
        while not done:
            # make a copy of the current array
            intermediate_array = [a[:] for a in self.current]
            # run update command
            self.update()
            # check if done
            done = (intermediate_array == self.current)

def answer(population, x, y, strength):
    # make simulation object
    my_sim = simulation(population, x, y, strength)
    # return final result
    return my_sim.current
````

With the zombie cure injections ready to go, it's time to start treating our zombified rabbit friends (known as zombits) at our makeshift zombit treatment center. You need to run out really fast to buy some gauze pads but you only have 30 seconds before you need to be back.

Luckily, the corner store has unlimited gauze pads in squares of all sizes. Jackpot! The pricing is simple - a square gauze pad of size K x K costs exactly K * K coins. For example, a gauze pad of size 3x3 costs 9 coins.

You're in a hurry and the cashier takes a long time to process each transaction. You decide the fastest way to get what you need is to buy as few gauze pads as possible, while spending all of your coins (you can always cut up the gauze later if you need to). Given that you have n coins, what's the fewest number of gauze pads you can buy?

Write a method answer(n), which returns the smallest number of square gauze pads that can be bought with exactly n coins.

n will be an integer, satisfying 1 <= n <= 10000.

````python
import math

def answer(n):
    '''
    the fewest number of gauze pads you can buy given that you have n coins
    '''
    # number of gauze pads
    num = 0
    # we're not done yet
    done = False
    while not done:
        # add to the number of pads needed
        num += 1
        # calculate the largest square less than or equal to n
        m = int(math.sqrt(n))**2
        # if m is n we're done
        if m == n:
            done = True
        n -= m
    
    return num
````

````python
# First attempt was a total failure
# answer from https://github.com/WilbertHo/foobar/tree/master/when_it_rains_it_pours
def answer(heights):
    '''
    We need to work our way in from both sides
    keeping track of the max height on either side
    '''
    
    left = 0 
    left_max = 0
    right = len(heights)-1
    right_max = 0
    
    water = 0
    while left < right:
        # update maxes
        if left_max < heights[left]:
            left_max = heights[left]
        if right_max < heights[right]:
            right_max = heights[right]
        
        if right_max >= left_max:
            # if right is higher that means the water will fill up
            # the left bin to left max, and we should move the left
            # index
            water += left_max - heights[left]
            left+=1
        else:
            # if the left is heigher we should move the other side.
            water += right_max - heights[right]
            right-=1
            
    return water
````

````python
# This is a job for recursion!
# The recusive solution wasn't accepted
def find_all(chunk, word):
    '''
    Find all matches of chunk and word
    
    Return a list of chunks with word removed
    '''
    
    word_len = len(word)
    list_removed = []
    
    # special cases
    if len(chunk) < len(word):
        return [chunk]

    if chunk == word:
        return ['']

    for i in range(len(chunk)-word_len+1):
        test_chunk = chunk[i:i+word_len]
        if test_chunk == word:
            list_removed.extend(
                find_all(chunk[:i]+chunk[i+word_len:], word)
            )
    
    # if no matches were found
    if len(list_removed) == 0:
        return [chunk]
    else:
        return list_removed

def answer_recursive(chunk, word):
    
    return sorted(find_all(chunk, word))[0]
````

````python
# Had to move to a different type of exhaustive search
# a little less exhaustive, in this case we loop through
# the different positions of chunk and remove word at that
# position, if it exists there. Then continue stripping out
# the words

# inspiration from: https://github.com/rtheunissen/foobar/blob/master/string_cleaning.py
def remove_sub(chunk, i, length):
    return chunk[:i]+chunk[i+length:]

def remove_all(chunk, word, offset):
    
    word_len = len(word)
    i = chunk.find(word, offset)

    while i > -1:
        chunk = remove_sub(chunk, i, word_len)
        i = chunk.find(word)
    return chunk

def answer_linear(chunk, word):
    
    if chunk == '' or chunk == word:
        return chunk
    
    all_possibilities = [remove_all(chunk, word, i) for i in range(len(chunk)-len(word))]
        
    if len(all_possibilities):
        return sorted(all_possibilities)[0]
    else:
        return ''

answer = answer_linear
````

Minion's bored game
===================

There you have it. Yet another pointless "bored" game created by the bored minions of Professor Boolean.

The game is a single player game, played on a board with n squares in a horizontal row. The minion places a token on the left-most square and rolls a special three-sided die. 

If the die rolls a "Left", the minion moves the token to a square one space to the left of where it is currently. If there is no square to the left, the game is invalid, and you start again.

If the die rolls a "Stay", the token stays where it is. 

If the die rolls a "Right", the minion moves the token to a square, one space to the right of where it is currently. If there is no square to the right, the game is invalid and you start again.

The aim is to roll the dice exactly t times, and be at the rightmost square on the last roll. If you land on the rightmost square before t rolls are done then the only valid dice roll is to roll a "Stay". If you roll anything else, the game is invalid (i.e., you cannot move left or right from the rightmost square).

To make it more interesting, the minions have leaderboards (one for each n,t pair) where each minion submits the game he just played: the sequence of dice rolls. If some minion has already submitted the exact same sequence, they cannot submit a new entry, so the entries in the leader-board correspond to unique games playable. 

Since the minions refresh the leaderboards frequently on their mobile devices, as an infiltrating hacker, you are interested in knowing the maximum possible size a leaderboard can have.

````python
def answer(t, n):
    '''
    we need to walk the tree starting from the bottom and
    only counting paths that END on the right
    
    copied from: https://github.com/huntzhan/foobar/blob/master/minions_bored_game.py
    
    Foobar didn't like my version of searching all permuations.
    '''
    
    # set up positions
    board = [0]*(n+1)

    # place token at start, note the left most,
    # index = 0, is a dump for failures
    board[1] = 1
    
    # loop through die throws
    for throw in range(t):
        # we should only be moving tokens
        # that have a chance of reaching the end
        # we have t - throw left, so that means
        # the max distance we can be is
        begin = n - (t - throw)
        
        # even if we have a lot of throws we shouldn't be past
        # the left
        if begin < 1:
            begin = 1
        
        # make a new board
        new_board = [0]*(n+1)
        # save the number of paths that have
        # made it to the right
        new_board[n] = board[n]
        # loop through positions begin to right-1
        # remember "you cannot move left or right from the rightmost square"
        for i in range(begin, n):
            # move tokens (paths) left
            new_board[i-1] += board[i]
            # tokens stay
            new_board[i] += board[i]
            # tokens move right
            new_board[i+1] += board[i]
        
        # update board
        board = new_board
        
    # return number of successful paths modulo
    return board[n]
````

````python
def rec(t, n):
    return (t - 2*(n-1))*n*3

def top_tri(t, n):
    return (n-2)*(n-3)*3//2 + (n-1)*2 + (n-2)*2

def bottom_tri(t, n):
    return (n-1)*(n-2)*3//2 + (n-1)*2

def intersection(t, n):
    t0 = 2*(n - 1)
    return (n + (n-1) + (n-2))*(t0-t)

def sum_all(t, n):
    return rec(t, n) + top_tri(t, n) + bottom_tri(t, n) + intersection(t, n)
````

````python
def k(t, n):
    if n-1 > t:
        return 0
    elif n-1 == t:
        return n
    else:
        return k(t-1, n) - k(t-2,n)*(t-n) + k(t-3,n)*(t-n+1)
````
